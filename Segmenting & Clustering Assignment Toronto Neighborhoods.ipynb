```python
# import random # library for random number generation
import numpy as np # library for vectorized computation
import pandas as pd # library to process data as dataframes

import matplotlib.pyplot as plt # plotting library
import requests #html query library
from bs4 import BeautifulSoup #html scraping library
```

```python
res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M") #Saves html link
soup = BeautifulSoup(res.content,'lxml') #Saves an element of that link (lxml) to be scraped
table = soup.find_all('table')[0]  #Scrapes all data in that element into a table 
df = pd.read_html(str(table), header=0)[0] #Turns that table into a dataframe
df.head(25)
```

```python
df = df.drop(df[df["Borough"] == 'Not assigned'].index) #Drops cells with a borough that is Not assigned.
df.head(25)
```

```python
df['Neighbourhood'] = np.where(df['Neighbourhood'] == 'Not assigned', df['Borough'], df['Neighbourhood']) #If a cell has a borough but a Not assigned neighbourhood, replaces it with the value of the borough. 
df.head(25)
```

```python
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index() #Combines multiple neighbourhoods under same postal code. 
df.shape #Prints number of rows/columns
```
